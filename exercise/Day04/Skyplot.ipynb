```python
import sys
import os
import math
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.fftpack import fft, ifft
from mpl_toolkits.mplot3d import Axes3D
```

```python
def skyplot(alpha,phi,f):
    A,D=np.meshgrid(alpha,phi)
    X=np.sin(D)*np.cos(A)
    Y=np.sin(D)*np.sin(A)
    Z=np.cos(D)
    output=np.zeros((len(phi),len(alpha)))
    for j in range(len(alpha)):
        for i in range(len(phi)):
            output[i][j]=f[i][j]

    fig,ax=plt.subplots(subplot_kw={"projection":"3d"})
    surf=ax.plot_surface(X,Y,Z,facecolors=plt.cm.viridis(np.abs(output)))
    
    plt.show()
def F_plus(alpha,phi):
    A,B=np.meshgrid(alpha,phi)
    return 0.5*(1+np.cos(B)**2)*np.cos(2*A)
def F_cross(alpha,phi):
    A,B=np.meshgrid(alpha,phi)
    return np.cos(B)*np.sin(2*A)
alpha=np.arange(0,2*pi,pi/600)
phi=np.arange(0,pi,pi/600)
f1=F_cross(alpha,phi)
f2=F_plus(alpha,phi)
skyplot(alpha,phi,f1)

skyplot(alpha,phi,f2)    
```
