```python
A=10
ta=3
f0=20
f1=30
phi0=pi/10
L=5
def s(A,ta,f0,f1,phi0,t):
    seq=np.zeros(len(t))
    for i in np.arange(0,len(t)-1):
        if (t[i]<=ta+L)and(t[i]>=ta):
            seq[i]=1
        else:
            seq[i]=0
    return A*np.sin(2*pi*(f0*(t-ta)+f1*(t-ta)**2)+phi0)
s(A,ta,f0,f1,phi0,t)
```
