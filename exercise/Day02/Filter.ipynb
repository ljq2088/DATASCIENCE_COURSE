```python
import sys
import os
import math
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.fftpack import fft, ifft
from scipy.signal import firwin,lfilter
```

```python
A1=10
A2=5
A3=2.5
f1=100
f2=200
f3=300
phi1=0
phi2=pi/6
phi3=pi/4
```

```python
def apply_lowpass_filter(signal, cutoff_freq, sampling_freq, filter_order):
    # 计算滤波器的系数
    nyquist_freq = 0.5 * sampling_freq
    normalized_cutoff_freq = cutoff_freq / nyquist_freq
    filter_coefficients = firwin(filter_order, normalized_cutoff_freq)

    # 应用滤波器
    filtered_signal = lfilter(filter_coefficients, 1.0, signal)

    return filtered_signal
sampling_freq=1024
sig_len=2048
delta=1/sampling_freq
t=np.arange(sig_len)/sampling_freq
signal=A1*np.sin(2*pi*f1*t+phi1)+A2*np.sin(2*pi*f2*t+phi2)+A3*np.sin(2*pi*f3*t+phi3)
cutoff_freq=150
filter_order=30
filtered_signal=apply_lowpass_filter(signal,cutoff_freq,sampling_freq,filter_order)
plt.figure()
plt.plot(t, signal, label='Input Signal')
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
